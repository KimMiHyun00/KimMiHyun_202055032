```python
import numpy as np
```

```python
"""
Introduction Broadcasting
>>차원이 다른 배열간 이항연산, 배열-스칼라 연산 가능

"""

a = np.ones((3,3))
b = np.arange(3)

print(a+b) #b가 브로드캐스팅이 되어 차원이 다른 두 배열이 더해짐
```

```python
# Broadcast both arrays
a = np.arange(3)
# print(a.ndim)
a = a[:,np.newaxis] #np.newaxis: 단순히 배열 차원 추가
# print(a.ndim)
print(a)
b = np.arange(3)
print(b)

print('\n')
print(a+b)
```

```python
# Broadcasting Imcompatible case
a = np.ones((3,2))
b = np.arange(3)
print(a + b)
```

```python
"""
Compatsion, Masks, and Boolean Logic
"""

x = np.array([1,2,3,4,5])
print(x != 3)
```

```python
# Bool masking 

np.random.seed(0)
a = np.random.randint(10,size=(3,4))
print(a)
print(a < 6)
```

```python
# boolean logic
np.count_nonzero(a<6) #True :1, False: 0
np.sum(a<6)

np.any(a>8)
np.all(a>1)
```

```python
"""
Masking Operation
"""

print(a)
print(a<5)
print(a[a<5]) #masking opeeration
```

```python
"""
Fancy Indexing
"""

np.random.seed(1)
x = np.random.randint(100,size=10)
print(x)

# fancy indexing
ind = [2,4,6]
print(x[ind])

# basic 
a = x[2]; b = x[4]; c = x[6]
print(a,b,c)
```

```python
# For n-D arrays

np.random.seed(2)
x = np.random.randint(10,size=(3,5))
print(x)

row = [0,1,2]
col = [2,1,4]

print(x[row,col])

```

```python
# fancy indexing with broadcasting
print(x)

# list->ndarray
row = np.array([0,1,2])
col = np.array([2,1,4])
print(x[row[:,np.newaxis],col])
```

```python
"""
Combined Indexing
"""

x = np.arange(1,13).reshape(3,4)

# print(x[2,[2,0,1])
ind = [2,0,1]

print(x[2], ind)
```

```python
"""
Example: Selectiing Random Points
"""
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set() # set plot styles
%config InlineBackend.figure_format = 'svg'

mean = [0,0]
cov = [[1,2],
      [2,5]]

rand = np.random.RandomState(1)
x = rand.multivariate_normal(mean,cov,100)
print(x)
print(x.shape)
```

```python
plt.scatter(x[:,0],x[:,1]);
```

```python
ind = np.random.choice(x.shape[0],20,replace False)
print(ind)
```

```python
selected = x[ind]
print(selected)
print(selected.shape)
```

```python
plt.scatter(x[:,0],x[:,1],alpha=0.3)
plt.scatter(selected[:,0],selected[:,1],facecolor='red',s=50)
```
